```
import spacy

nlp = spacy.blank('en')

doc = nlp('SpaCy is an Industrial-Strength Natural Language Processing library designed to help you do real work — to build real products, or gather real insights.')

libraries = [doc[0]]
nouns = [doc[9], doc[16], doc[21], doc[26]]

adjectives = [doc[15], doc[20], doc[24]]

verbs = [doc[10], doc[12], doc[14], doc[19], doc[24]]

spacy_description = doc[0:10]
```

```
nlp = spacy.load('en_core_web_sm')

doc = nlp('No matter how many times you die, be always sure that you\'re reborn as yourself.')

nouns = []
pronouns = []
adjectives = []
verbs = []
others = []

for token in doc:
    if token.pos_ == 'NOUN':
        nouns.append(token)
    elif token.pos_ == 'ADJ':
        adjectives.append(token)
    elif token.pos_ == 'VERB':
        verbs.append(token)
    elif token.pos_ == 'PRON':
        pronouns.append(token)
    else:
        others.append(token)

print('NOUNS:', nouns)
print('PRONOUNS: ', pronouns)
print('ADJECTIVES: ', adjectives)
print('VERBS: ', verbs)
print('OTHERS: ', others)
```

```
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import plotly.express as px
from PIL import Image
import pandas as pd

%matplotlib inline

data = {
    'nouns': nouns,
    'pronouns': pronouns,
    'adjectives': adjectives,
    'verbs': verbs,
    'others': others
}

lorem_ipsum = '''
Many times you might have seen a cloud filled with lots of words in different sizes, which represent the frequency or the importance of each word. This is called Tag Cloud or WordCloud. For this tutorial, you will learn how to create a WordCloud of your own in Python and customize it as you see fit. This tool will be quite handy for exploring text data and making your report more lively.
'''

wordcloud = WordCloud(
    max_font_size=80, 
    max_words=100, 
    background_color='white', 
    repeat=False, 
    width=1280, 
    height=720
    ).generate(lorem_ipsum)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear', aspect='auto')
plt.axis("off")
plt.show()
```
